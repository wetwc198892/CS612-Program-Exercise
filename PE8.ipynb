```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Heights.csv')
plt.scatter(x = data['Husband'],y = data['Wife'])
plt.xlabel('Husband')
plt.ylabel('Wife')
plt.title('Scatter Plot of Heights')
x_lim=plt.xlim()
y_lim = plt.ylim()
plt.plot(x_lim,y_lim,'k-')
plt.xlim(x_lim)
plt.ylim(y_lim)
plt.show()
```

```python
x = np.arange(-5,5,step=0.1)
```

```python
y = -(1/3)*x+1.5
plt.plot(x,y,color='g',linestyle=':')
plt.legend(['Graph1'],loc=9)
```

```python
y = 0.02 * x * (x-1) * (x+1) + 1
plt.plot(x,y,color='y',linestyle='--')
plt.legend(['Graph2'],loc=9)
```

```python
Y = 2 * np.exp(-x**2)
plt.plot(x,Y,color='r',linestyle='-')
plt.legend(['Graph3'],loc=9)
```
